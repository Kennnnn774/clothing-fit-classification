```python
import gzip
import json
import random
import numpy as np
import pandas as pd
```

```python
def reading_gz(filename):
    f = gzip.open(filename, 'rt', encoding="utf8")
    lines = []
    for line in f:
        liness = json.loads(line.strip())
        lines.append(liness)
    return lines
```

```python
def category_identity (category):
    category = stemmer.stem(category)
    poncho = ['poncho','crewneck','cape']
    coat = ['coat','peacoat','overcoat','suit','blazer','bomber','jacket','parka','tunic']
    tshirt = ['t-shirt','tee','shirt','shift','shirtdress','skirt','skort','top','vest','blous','blous','blouson','cami','mini']
    sweater = ['sweater','sweatershirt','sweatshirt','caftan','cardigan','henley','hoodi','midi','pullov','blou']
    culott = ['culott','frock','jumpsuit','ballgown','buttondown','combo','dress','gown','kaftan',
              'kimono','maxi','romper','sheath','turtleneck']
    overal = ['overal','duster','knit','print','tank','trench','for','over']
    down = ['down','jean','jogger','leg','pant','sweatpant','tight','trouser',]
    if category in poncho:
        return 'poncho'
    if category in coat:
        return 'coat'
    if category in tshirt:
        return 'tshirt'
    if category in sweater:
        return 'sweater'
    if category in culott:
        return 'culott'
    if category in overal:
        return 'overal' 
    if category in down:
        return 'down' 
    return category
```

```python
data_tb['catgory'] = data_tb.category.apply(category_identity)
```

## update the weight

```python
data_tb['size'].mean()
```

```python
#convert to weight to float then minus the average of weight
def convert_weight(strin):
    if isinstance(strin, str):
        return (float(strin.strip('lbs')) -137) *0.4535
    else:
        return (float(strin) - 137) *0.4535
    
#convert to height to cm then minus the average of height
def convert_height(strin):
    if isinstance(strin, str):
        #print(strin)
        fet, inc = strin.split()
        fet = float(fet.strip("'").strip('"'))
        inc = float(inc.strip("'").strip('"'))
        return (fet * 30.48+ inc * 2.54) - 165.89

def normalize_the_size(strin):
    return strin - 12.245175128801728
#fillna with the average - average, which is 0
data_tb.weight = data_tb.weight.apply(convert_weight).fillna(0)
data_tb.height = data_tb.height.apply(convert_height).fillna(0)

#change the size to float
data_tb['size'] = data_tb['size'].astype(float).apply(normalize_the_size)
```

since we think the height people tend to have a higher weight and bigger size of dress, so we add those three value together.

```python
data_tb['size_guess_fit'] = data_tb['weight'] + data_tb['height'] + data_tb['size']
```

Now we try to fix the bust size. We decide to spilt it into two column, which is using the number as one and use the character following as the other. Moreover, we fillna with the 34 dd as I google search the average women's bust size in US.

```python
data_tb['bust size'] = data_tb['bust size'].fillna('34dd')
```

```python
def get_bust_num (data):
    return int(data[:2])
def get_bust_character(data):
    return data[2:]
```

```python
data_tb['bust_num'] = data_tb['bust size'].apply(get_bust_num)
data_tb['bust_char'] = data_tb['bust size'].apply(get_bust_character)
```
